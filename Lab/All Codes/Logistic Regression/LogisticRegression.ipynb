```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv('Logistic.csv')
print("The dataset is")
dataset
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print("The samples are :")
print(X)
print("The target sample is :")
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()
regressor.fit(X_train,y_train)
print("Training the ML Model")
```

```python
print(regressor.predict(X_test))
print(y_test)
```

```python
print("The score of the dataset")
regressor.score(X_test,y_test)
```

```python
m=regressor.coef_
c=regressor.intercept_
print("Intercept is ")
print(m)
print("Value of C is ")
print(c)
```

```python
import math
def sigmoid(m,c,x):
  return 1/(1+math.exp(-(m*x+c)))
```

```python
print("The probability of passing for a student studying 33hr")
print(sigmoid(m,c,33))
```

```python
viz_train=plt
viz_train.scatter(X_train,y_train,color="red")
viz_train.xlabel("Hours Studies")
viz_train.ylabel("Result(Probability)")
print("Plot of the model")
viz_train.show()
```

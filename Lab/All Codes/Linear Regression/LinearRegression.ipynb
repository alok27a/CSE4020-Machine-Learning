```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('student_scores.csv')
dataset
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

**Fitting Simple Linear Regression to the training set**

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

**Visualising of Training and testing model**

```python
viz_training=plt
viz_training.scatter(X_train,y_train,color='blue')
viz_training.plot(X_train,regressor.predict(X_train),color='red')
viz_training.xlabel("Hours")
viz_training.ylabel("Scores")
viz_training.show()
```

```python
viz_testing=plt
viz_testing.scatter(X_test,y_test,color='blue')
viz_testing.plot(X_test,regressor.predict(X_test),color='red')
viz_testing.xlabel("Hours")
viz_testing.ylabel("Scores")
viz_testing.show()
```

**Value of Coeff and Intercept**

```python
regressor.coef_
```

```python
regressor.intercept_
```

**Prediction of Values**

```python
regressor.predict([[6]])
```

**Accuracy**

```python
regressor.score(X_test,y_test)
```

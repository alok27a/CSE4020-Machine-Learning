# Linear Regression Example

```python
import numpy as np
import pandas as pd

dataset=pd.read_csv('q2-2.csv')
print("Dataset is :")
dataset
```

**X Values(Independent Values) and Y Values(Dependent Values)**

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print("The training sample is : ")
print(X)
print("The target concept is :")
print(y)
```

**Training of Data**

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)

from sklearn.linear_model import LinearRegression

regressor=LinearRegression()
print("Training the ML model ")
regressor.fit(X_train,y_train)
```

**Lets Visualise our training models and testing models**

```python
import matplotlib.pyplot as plt

viz_train=plt
viz_train.scatter(X_train,y_train,color="red")
viz_train.plot(X_train,regressor.predict(X_train),color="blue")
viz_train.xlabel("Years of Experience")
viz_train.ylabel("Salary")
viz_train.show()
```

```python
m=regressor.coef_
c=regressor.intercept_
print("The linear equation is :")
print("y=0.66*x + 0.33")
```
